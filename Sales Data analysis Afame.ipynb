# Task 2. Compute Total Sales

```python
total_sales = data['Sales'].sum()
print(f'Total Sales: {total_sales:,.2f}')
```

# Task 3. Analyze Sales Trends Over Time

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month
data['Quarter'] = data['Order Date'].dt.to_period('Q')
monthly_sales = data.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
quarterly_sales = data.groupby(['Year', 'Quarter'])['Sales'].sum().reset_index()
yearly_sales = data.groupby(['Year'])['Sales'].sum().reset_index()
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Month'], monthly_sales['Sales'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

# task 4 . determine the best-selling products, and build visualizations.

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
product_sales = data.groupby(['Product ID', 'Product Name'])['Sales'].sum().reset_index()
best_selling_product = product_sales.loc[product_sales['Sales'].idxmax()]
```

```python
print(f"Best Selling Product: {best_selling_product['Product Name']} with Sales: {best_selling_product['Sales']}")
```

```python
top_products = product_sales.nlargest(10, 'Sales')
plt.figure(figsize=(12, 8))
sns.barplot(x='Sales', y='Product Name', data=top_products)
plt.title('Top 10 Best Selling Products')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.show()
```

# Sales Distribution of best Selling product  By Segment 

```python
best_product_data = data[data['Product ID'] == best_selling_product['Product ID']]
plt.figure(figsize=(7, 4))
sns.barplot(x='Sales', y='Segment', data=best_product_data, estimator=sum)
plt.title(f'Sales Distribution of {best_selling_product["Product Name"]} by Segment')
plt.xlabel('Total Sales')
plt.ylabel('Customer Segment')
plt.show()
```
